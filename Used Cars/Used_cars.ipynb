# Handeling Missing And Unwanted Data

```python
df['Mileage']=pd.to_numeric(df['Mileage'].str.split().str.get(0),errors='coerce')
df['Engine']=pd.to_numeric(df['Engine'].str.split().str.get(0),errors='coerce')
df['Power']=pd.to_numeric(df['Power'].str.split().str.get(0),errors='coerce')
df.head(2)
```

```python
df.drop(['New_Price','S.No.'],axis=1,inplace=True)
df.head(2)
```

```python
df.fillna(value={'Engine':df['Engine'].median(),'Power':df['Power'].median(),'Seats':df['Seats'].median(),'Price':df['Price'].median()},inplace=True)
```

```python
df.dropna(inplace=True)
```
